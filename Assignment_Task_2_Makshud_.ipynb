```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits


digits = load_digits()

# Features: 64 pixel-intensity values
X = digits.data
# Labels: digits 0 to 9
y = digits.target

# Here I print the information which is the features of the shape
print("Shape of feature matrix:", X.shape)  # Expect (1797, 64)
print("Shape of labels:", y.shape)         # Expect (1797,)

# Then It shows an example image
plt.imshow(digits.images[0], cmap='grey')
plt.title(f"Example digit: {digits.target[0]}")
plt.colorbar()
plt.show()

from sklearn.decomposition import PCA

# Reduce to 2 principal components for easy plotting
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)  # shape: (1797, 2)

# Here I Plot the PCA results in 2D
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y,
    cmap='Spectral',
    s=10
)
plt.colorbar(scatter, label='Digit Label')
plt.xlabel("Principal Comp 1")
plt.ylabel("Principal Comp 2")
plt.title("PCA on Digits Dataset")
plt.show()


print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained by these two components:",
      np.sum(pca.explained_variance_ratio_))
```
